   \Rightarrow k &= \frac{F}{x}
 \end{aligned}
 
 And our data set is of the $(F_i, x_i)$ pairs, therefore $-k$ can be obtained by the line of bestfit when $F$ is plotted against $x$

```python
F = [k[0] for k in complete_data_set]
x = [k[1] for k in complete_data_set]
```

Now that we have our $F$ and $x$ lists, we can plot  $F$ against $x$

```python
pylab.plot(x, F, 'b.')
```

```python
# Get the line of best fit. $m_b$ is the best slope.
m_b, b_b = pylab.polyfit(x, F, 1)
k = m_b*-1
print "Best Slope: ", m_b
print "Best Intercept: ", b_b
print "Spring constant: ", k_b
```

```python
def calculate_intercept(m, (y, x)):
    """
    Calculates the intercept (given a point and slope)
    """
    return (y - m*x)
```

```python
# find the maximum and the minimum slopes ($a$ and $b$)
a_X_min = min(x[-250:-1])
a_X_max = max(x[0:250])
a_F_min = F[x.index(a_X_min)]
a_F_max = F[x.index(a_X_max)]
a = float(a_F_max - a_F_min)/(a_X_max - a_X_min)
a_i = calculate_intercept(a, (a_F_max, a_X_max))
print "Slope 1:", a
print "Intecept 1:", a_i

b_X_min = max(x[-250:-1])
b_X_max = min(x[0:250])
b_F_min = F[x.index(b_X_min)]
b_F_max = F[x.index(b_X_max)]
b = float(b_F_max - b_F_min)/(b_X_max - b_X_min)
b_i = calculate_intercept(b, (b_F_max, b_X_max))
print "Slope 2:", b
print "Intecept 2:", b_i
```
